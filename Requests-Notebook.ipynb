# Spotipy Request Chain
### Goal: Use the Spotipy API library in python to pull song lyrics from Canadas top playlist.

```python
import sys
sys.path.append('../')

import os

from src.SpotipyAPI import *
from src.azlyrics import *
```

```python
# Grab Spotify dev keys from environment variables
client_id = os.getenv('SPOTIFY_CLIENT_ID')
client_secret = os.getenv('SPOTIFY_CLIENT_SECRET')
```

```python
# Initialize Class
spotipy = SpotifyAPI(client_id, client_secret)
```

```python
# Initialize API URL
URI = 'https://api.spotify.com/v1'

playlist_id = '37i9dQZF1DWXT8uSSn6PRy' #Hot Hits Canada

playlist_tracks = f'/playlists/{playlist_id}/tracks'

url_main = URI+playlist_tracks
print(url_main)
```

```python
json_res = get_json(url_main, spotipy)
```

```python
json_res['items'][0]
```

```python
len(json_res['items'])
```

```python
# Create Dataframe
df = pd.DataFrame(columns=['Artist', 'Song', 'Lyrics'])
```

```python
# Assign artist and song data for each track to new dataframe
for i in tqdm(range(len(json_res['items']))):
    df.at[i, 'Artist'] = json_res['items'][i]['track']['album']['artists'][0]['name']
    df.at[i, 'Song'] = json_res['items'][i]['track']['name']
```

```python
df
```

```python
#Assign Lyrics to dataframe using get_lyrics function from azlyrics.py
for i in tqdm(range(len(df))):
    df.at[i, 'Lyrics'] = get_lyrics(df.iloc[i, 0], df.iloc[i, 1])
```
